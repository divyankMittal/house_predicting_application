```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('delhi.csv')

df['Status'] = df['Status'].astype('category')
df['Status'] = df['Status'].cat.codes

df['neworold'] = df['neworold'].astype('category')
df['neworold'] = df['neworold'].cat.codes

df['Furnished_status'] = df['Furnished_status'].astype('category')
df['Furnished_status'] = df['Furnished_status'].cat.codes


df['type_of_building'] = df['type_of_building'].astype('category')
df['type_of_building'] = df['type_of_building'].cat.codes

df['Balcony'] = df['Balcony'].astype('category')
df['Balcony'] = df['Balcony'].cat.codes

df['parking'] = df['parking'].astype('category')
df['parking'] = df['parking'].cat.codes

df['Lift'] = df['Lift'].astype('category')
df['Lift'] = df['Lift'].cat.codes

X = df.drop(columns = 'price')
Y = df['price']


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=0)
lr = LinearRegression()

lr.fit(X_train,Y_train)

```

```python
Y_pred_train = lr.predict(X_test)


import matplotlib.pyplot as plt
plt.scatter(Y_test, Y_pred_train)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
```

```python
Y_pred_train = lr.predict(X_test)
from sklearn import metrics
metrics.r2_score(Y_train, Y_pred_train)

import matplotlib.pyplot as plt
plt.scatter(Y_train, Y_pred_train)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('delhi.csv')

df['Status'] = df['Status'].astype('category')
df['Status'] = df['Status'].cat.codes

df['neworold'] = df['neworold'].astype('category')
df['neworold'] = df['neworold'].cat.codes

df['Furnished_status'] = df['Furnished_status'].astype('category')
df['Furnished_status'] = df['Furnished_status'].cat.codes


df['type_of_building'] = df['type_of_building'].astype('category')
df['type_of_building'] = df['type_of_building'].cat.codes

df['Balcony'] = df['Balcony'].astype('category')
df['Balcony'] = df['Balcony'].cat.codes

df['parking'] = df['parking'].astype('category')
df['parking'] = df['parking'].cat.codes

df['Lift'] = df['Lift'].astype('category')
df['Lift'] = df['Lift'].cat.codes

X = df.drop(columns = 'price')
Y = df['price']


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=0)
lr = LinearRegression()

lr.fit(X_train,Y_train)
```
